Scatter plot
- plot and show the relationship for 2 sets of value
- and see if there are correlated

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use("seaborn")
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5] 
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
          538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolors='black', linewidth=1, alpha=0.75)
plt.savefig('figure/satisfaction_level.png')

cbar = plt.colorbar()
cbar.set_label('Satisfaction level')
```

Visualize the correlation between the view count, total likes and the like/dislike ratio.

```python
 data = pd.read_csv('2019-05-31-data.csv')
 data.head(5)
 view_count = data['view_count']
 likes = data['likes']
 ratio = data['ratio']


```

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', 
            edgecolors='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Likes/Dislikes ratio')
plt.xscale('log') # make outlier looks better
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.savefig('figure/views_likes_ratio.png')
plt.show()

```
