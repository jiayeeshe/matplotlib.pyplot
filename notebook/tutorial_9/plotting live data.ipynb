```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
from itertools import count

# Create a figure
# fig, ax = plt.subplots()
x_vals = []
y_vals = []
plt.plot()


index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))

    plt.cla()  # Use ax.clear() instead of plt.cla()
    plt.plot(x_vals, y_vals, color='red', label='line1')
    plt.xlabel('x starts from 1')
    plt.ylabel('random y value from 0 to 5')
    plt.legend(loc='upper right')


ani = FuncAnimation(plt.gcf(), animate, interval=1000, cache_frame_data=False)  # Pass fig instead of plt.gcf()

plt.tight_layout()
plt.show()
```
