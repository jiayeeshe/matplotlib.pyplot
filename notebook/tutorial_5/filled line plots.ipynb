Filled line plots
- Enhance the line plots visualization
- Helps in add on third level information

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data.head(5)
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median_python = 57287 # y2 value
plt.fill_between(ages, py_salaries, dev_salaries,
                where=(py_salaries>dev_salaries),
                interpolate=True, alpha=0.25, label='Above Avg')
plt.fill_between(ages, py_salaries, dev_salaries,
                where=(py_salaries<=dev_salaries),
                interpolate=True, color='red', alpha=0.25, label='Below Avg')  # y2 mention where to fill until
plt.legend()
plt.title("Median salaries (USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

```
